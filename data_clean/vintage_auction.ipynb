```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv('https://bit.ly/vintage-auctions')
```

```python
df.columns = df.columns.str.replace(' ', '_').str.lower()
```

```python
for col in df.columns[3:]:
  print(col)
  print(df[col].value_counts(dropna=False))
  print()
  print()
```

```python
df.head(2)
```

```python
df = (
    df.assign(
        time = pd.to_datetime(df.time),
        auction = df.auction.apply(lambda x: x.replace('$', '')).astype(float),
    )
)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.color.fillna('Unknown', inplace=True)
```

```python
df.condition.fillna('Unknown', inplace=True)
```

```python
df.condition = df.condition.apply(lambda x: x.title())
```

```python
df.condition.value_counts()
```

```python
df.isna().sum()
```

```python
df.time.min(), df.time.max()
```

```python
df.auction.min(), df.auction.max()
```

```python
df['day_period'] = pd.cut(df.time.dt.hour, bins=[5, 12,18, 24 ], labels=['Mor', 'Af', 'Ev'])
```

```python
df.head()
```

```python
df.groupby(['day_period', 'condition']).mean()['auction']
```

```python
df2 = pd.get_dummies(df[['color', 'condition', 'day_period']])
df2.head()
```

```python
df = pd.concat([df.drop(columns=['color', 'condition', 'day_period']), df2], axis=1)
```

```python
df.head()
```

```python
df.shirt_size.unique()
```

```python
size2id = {
    'XS': 0,
    'S': 1,
    'M': 2,
    'L': 3,
    'XL': 4,
    '2XL': 5,
    '3XL': 6,
}
df.shirt_size = df.shirt_size.map(size2id)
```

```python
df3 = pd.get_dummies(df['shirt_size'])
df = pd.concat([df.drop(columns=['shirt_size']), df3], axis=1)
```
