## Projet 1 : Tracé de l'équation d'Einstein

### Partie 1 : Création de données à partir d'une équation

Il est important de pouvoir traduire directement une équation réelle en un diagramme/graphique. Votre première tâche est vraiment du pure NumPy, puis nous allons explorer comment la tracer avec Matplotlib. [L'équation mondialement connue](https://fr.wikipedia.org/wiki/E%3Dmc2) de Einstein:

$$E=mc^2$$


Utilisez vos connaissances de NumPy pour créer deux tableaux : E et m, où **m** correspond simplement à 11 valeurs uniformément espacées représentant 0 à 10 grammes. E doit être l'énergie équivalente à la masse. Vous devrez déterminer ce qu'il faut fournir à **c** avec les unités m/s, une recherche rapide sur Google vous donnera facilement la réponse (nous utiliserons une approximation dans nos solutions).

**NOTE : Si cela vous perturbe, passez à la vidéo des solutions pour une correction guidée.**

```python
c = 299_792_458
m = np.linspace(0, 10, 11)
E = m * c * c
```

### Partie 2 : Tracer E=mc^2

Maintenant que nous avons les tableaux E et m, nous pouvons tracer cela pour voir la relation entre Énergie et Masse.

```python
fig = plt.figure(figsize=(8, 5), dpi=85)
ax = fig.add_axes([0,0,1,1])
ax.plot(m, E, color='k', lw=2)
ax.set(xlabel='Mass(g)', ylabel='Energy(Joules)', title='E=mc²')
ax.set_yscale('log')
plt.show()
```

## Projet 2 : Création de tracés à partir de points de données

Dans le domaine financier, la courbe de rendement est une courbe montrant plusieurs rendements jusqu'à échéance ou taux d'intérêt sur différentes durées de contrat (2 mois, 2 ans, 20 ans, etc.) pour un contrat d'emprunt similaire. La courbe montre la relation entre le (niveau du) taux d'intérêt (ou coût d'emprunt) et le délai d'échéance, connu sous le nom de « terme », de la dette d'un emprunteur donné dans une devise donnée.

Les taux d'intérêt en dollars américains payés sur les titres du Trésor américain pour diverses échéances sont surveillés de près par de nombreux traders et sont généralement tracés sur un graphique tel que celui de droite, appelé officieusement « la courbe de rendement ».

La courbe de rendement (Yield Curve) est une représentation graphique de l'écart entre les taux d'intérêts des obligations d'État à court terme et à long terme, appelé « bons du Trésor ». En effet, tous les pays ont une courbe de rendement pour leurs marchés obligataires.

**Pour cet exercice, nous vous fournirons les données des courbes de rendement à deux moments distincts dans le temps. Nous vous demanderons ensuite de créer des tracés à partir de ces données.**



### Partie 1 : Données de courbe de rendement

Nous avons obtenu des données de courbe de rendement pour vous auprès du [département du Trésor américain](https://www.treasury.gov/resource-center/data-chart-center/interest-rates/pages/textview.aspx?data=yield). Les données montrent les intérêts payés pour une obligation du Trésor américain pour une certaine durée de contrat. La liste des étiquettes affiche la longueur de contrat correspondante par position d'index.

**Tâche : Exécutez la cellule ci-dessous pour créer les listes à tracer.**

```python
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']

july16_2007 =[4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
```

### Partie 2 : Tracé de graphiques
**Tâche : Déterminez comment tracer les deux courbes sur la même Figure. Ajoutez une légende pour montrer quelle courbe correspond à une année donnée.**

```python
fig = plt.figure(dpi=85)
ax = fig.add_axes([0,0,1,1])
ax.plot(labels, july16_2007, label='july16_2007')
ax.plot(labels, july16_2020, label='july16_2020')
ax.legend()
plt.show()
```

**Tâche : La légende du graphique ci-dessus semble un peu bizarre au milieu des courbes. Bien que cela ne soit pas bloquant, ce serait mieux s'il était en dehors du graphique. Découvrez comment déplacer la légende en dehors du graphique principal.**

```python
fig = plt.figure(dpi=85)
ax = fig.add_axes([0,0,1,1])
ax.plot(labels, july16_2007, label='july16_2007')
ax.plot(labels, july16_2020, label='july16_2020')
ax.set(xlabel='Years', ylabel='Interests')
ax.legend(loc=(1.05, 0.55))
plt.show()
```

**Tâche : Bien que le graphique ci-dessus montre clairement comment les taux ont baissé entre 2007 et 2020, les placer sur le même graphique rend difficile le discernement des différences de taux au cours de la même année. Utilisez `.subplots()` pour créer le graphique ci-dessous, qui montre la courbe de rendement de chaque année.**

```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(9, 5))
ax[0].plot(labels, july16_2007)
ax[1].plot(labels, july16_2020)
ax[0].set(xlabel='Years', ylabel='Interests', title='Jul, 16 2007')
ax[1].set(xlabel='Years', ylabel='Interests', title='Jul, 16 2020')
fig.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(dpi=105)

ax1.plot(labels, july16_2007, lw=2, color="blue")
ax1.set_ylabel(r"2007", fontsize=18, color="blue")
ax1.set_title('16 of july Interests', fontsize=15)
for label in ax1.get_yticklabels():
    label.set_color("blue")
    
ax2 = ax1.twinx()
ax2.plot(labels, july16_2020, lw=2, color="red")
ax2.set_ylabel(r"2020", fontsize=18, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")
```
