```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
y = 2 * x
```

```python
plt.figure(figsize=(8,5))
plt.plot(x, y, label='f(x)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('X Vs Y')
plt.legend()
plt.xlim(0, 10)
plt.ylim(0, 20)
plt.show();
```

```python
fig = plt.figure(figsize=(5, 5))
axs = fig.add_axes([0,0,1,1])
plt.plot(x, y, label='f(x)')

plt.legend()
plt.show();
```

```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
fig = plt.figure()
#bigg fig
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(a, b)
ax1.set_xlabel('A')
ax1.set_ylabel('B')

# small fig
ax2 = fig.add_axes([0.11, 0.5, 0.4, 0.4])
ax2.plot(a, b)
ax2.set_xlabel('A')
ax2.set_ylabel('B')
ax2.set_xlim(0, 1.5)
ax2.set_ylim(0,10)
ax2.set_title('Zoom')


plt.show()
```

```python
fig = plt.figure(dpi=85)
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_title('Quadratic')
axes1.set_xlim(0, 10)
axes1.set_ylim(0, 10000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
plt.savefig('quadratic.png', bbox_inches='tight');
```
