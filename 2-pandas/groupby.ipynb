```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("mpg.csv")
df.sample(8)
```

```python
df[['cylinders', 'horsepower', 'model_year', 'origin']].isna().sum()
```

```python
for col in ['cylinders', 'horsepower', 'model_year', 'origin', 'name']:
  print(col, ':')
  print(df[col].value_counts().sort_values(ascending=False))
  print()
```
