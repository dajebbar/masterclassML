```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

# Scatter plots

```python
fig = plt.figure(figsize=(12,5))
ax = fig.add_axes([0,0,1,1])
ax = sns.scatterplot(x='salary', y='sales', 
                     hue='level of education', data=df,
                     palette='Dark2'
                     )
plt.show()
```

```python
fig = plt.figure(figsize=(12,5))
ax = fig.add_axes([0,0,1,1])
ax = sns.scatterplot(x='salary', y='sales', 
                     s=200, data=df,
                     alpha=0.3
                     )
plt.show()
```

Most salaries between 60000 and 120000

```python
fig = plt.figure(figsize=(12,5))
ax = fig.add_axes([0,0,1,1])
ax = sns.scatterplot(x='salary', y='sales', 
                     s=200, data=df,
                     alpha=0.3, 
                     hue='level of education',
                     )
plt.show()
```

```python
fig = plt.figure(figsize=(12,5))
ax = fig.add_axes([0,0,1,1])
ax = sns.scatterplot(x='salary', y='sales', 
                     s=200, data=df,
                     style='level of education', 
                     hue='level of education',
                     )
plt.savefig('sales_salary.png', bbox_inches='tight')
```

# Distribution plots

Displays one continious feature
- rug plot
- histogram
- kde plot

```python
plt.figure(figsize=(8, 5))
sns.rugplot(x='salary', data=df, height=.12,)
plt.show()
```

```python
sns.displot(x='salary', data=df, bins=10, 
            color='#264653', edgecolor='#f4a261',
            kde=True,
            rug=True,
            )
plt.show()
```

```python
sns.kdeplot(x='salary', data=df, clip=[60000, 120000]);
```

# Categorical statistics plots

```python
plt.figure(figsize=(8,5))
plt.xticks(rotation=45)
sns.countplot(x='division', data=df)

plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.xticks(rotation=45)
sns.countplot(x='level of education', data=df)

plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.xticks(rotation=45)
sns.countplot(x='level of education', data=df,
              hue='division',
              palette='Dark2'
              )

plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.barplot(x='level of education', y='salary',
            data=df, estimator=np.mean,
            ci='sd'
            )
plt.show()
```

```python
df.groupby(['level of education']).agg(['mean', 'std'])[['salary']]
```

```python
plt.figure(figsize=(12, 5))
sns.barplot(x='level of education', y='salary',
            data=df, estimator=np.mean,
            ci='sd',
            hue='division'
            )
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.barplot(x='level of education', y='salary',
            data=df, estimator=np.mean,
            ci='sd',
            hue='division'
            )

plt.legend(bbox_to_anchor=(1.05, 0.55))
plt.show()
```

# Categorical Distribution plots

- boxplot
- violinplot
- swarmplot
- boxenplot (Letter-value plot)


```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
# boxplot with math score for all df

sns.boxplot(y='math score', data=df);
```

```python
#  boxplot with math score for test preparation course
sns.boxplot(y='math score', x='test preparation course', data=df);
```

```python
#  boxplot with reading score for test preparation course
sns.boxplot(y='reading score', x='test preparation course', data=df);
```

```python
#  boxplot with reading score for parental level of education	
plt.figure(figsize=(12, 5), dpi=85)
plt.xticks(rotation=45)
sns.boxplot(y='reading score', x='parental level of education', data=df);
```

```python
#  boxplot with reading score for parental level of education	
plt.figure(figsize=(12, 5), dpi=85)
plt.xticks(rotation=45)
sns.boxplot(x='reading score', y='parental level of education', data=df,
            hue='test preparation course',
            palette='Dark2'
            )
plt.legend(bbox_to_anchor=(1.05, 0.55));
```
