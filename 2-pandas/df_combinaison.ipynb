# Fusion: Merge

```python
import pandas as pd

registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})

print(registrations)
print()
print(logins)
```

```python
# inner
inner_df = pd.merge(registrations, logins, how='inner', on='name')
inner_df
```

```python
# left
left_df = pd.merge(registrations, logins, how='left', on='name')
left_df
```

```python
# right
right_df = pd.merge(registrations, logins, how='right', on='name')
right_df
```

```python
# outer
outer_df = pd.merge(registrations, logins, how='outer', on='name')
outer_df
```

```python
registrations.set_index('name', inplace=True)
registrations
```
